```python
# Import yfinance
import yfinance as yf

# Get the data for the stock ibm by specifying the stock ticker, start date, and end date
data1 = yf.download('IBM','2020-06-10','2020-06-13', interval='1m')
data1
```

```python
data2 = yf.download('IBM','2020-06-15','2020-06-20', interval='1m')
data2
```

```python
data3 = yf.download('IBM','2020-06-21','2020-06-27', interval='1m')
data3
```

```python
data4 = yf.download('IBM','2020-06-29','2020-07-04', interval='1m')
data4
```

```python
data5 = yf.download('IBM','2020-07-06','2020-07-09', interval='1m')
data5
```

```python
import pandas as pd
dataset=pd.concat([data1,data2,data3,data4,data5])
dataset
```
